```python
import pandas as pd
pd.set_option('display.max_colwidth', None)

united_df = pd.read_csv("all_crime_data.csv")
united_df.head()
```

```python
united_df.info()
```

```python
df_coord = united_df[["LAT_WGS84", "LONG_WGS84"]].copy()
df_coord["lat"] = round(df_coord["LAT_WGS84"], 3)
df_coord["lon"] = round(df_coord["LONG_WGS84"], 3)
df_coord_clean = df_coord[(df_coord["LAT_WGS84"] != 0) & (df_coord["LONG_WGS84"] != 0)].copy()
df_coord_clean.head()
```

```python
df_coord_clean["lat_long"] = df_coord_clean["lat"].astype(str) + ',' + df_coord_clean["lon"].astype(str)
df_coord_clean.head()
```

```python
df_coord_clean["lat_long"].nunique()
```

```python
df_coord_clean.drop(["LAT_WGS84", "LONG_WGS84"], inplace=True, axis=1)
df_coord_clean.head()
```

```python
df_coord_clean.drop_duplicates(inplace=True)
df_coord_clean.dropna(subset=['lat', 'lon'], how='all', inplace=True)
df_coord_clean.info()
```

```python
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent="big_data2")

from geopy.extra.rate_limiter import RateLimiter
geocode_reverse = RateLimiter(geolocator.reverse, min_delay_seconds=0.01)
```

```python
import re

def retrieve_zip_code(latitude: float, longitude: float) -> str:
    try:
        zip_code = geocode_reverse((latitude, longitude)).address.split(',')[-2]
        pattern = r'\s*([A-Za-z]\d[A-Za-z] \d[A-Za-z]\d)'
        match = re.search(pattern, zip_code)

        # Extract the postal code without the leading space
        if match:
            postal_code = match.group(1)
            return postal_code
        else:
            return None
    except:
        return None
```

```python
from tqdm import tqdm
tqdm.pandas()

df_coord_clean['zip_code'] = df_coord_clean.progress_apply(lambda row: retrieve_zip_code(row["lat"], row["lon"]), axis=1)

df_coord_clean.head()
```

```python
df_coord_clean.drop('lat_long', inplace=True, axis=1)
df_coord_clean.head()
```

```python
df_coord_clean.to_csv("coord_to_zip.csv", index=False)
```
