## Convert Prices to INT

```python
df = spark.sql('''SELECT *,
    CAST(REPLACE(SUBSTRING(price, 3, LENGTH(price) - 3), ',', '') AS INT) AS price_int
    FROM Lakehouse_BD2.house_housing_zipcode''')
```

```python
df.write.mode("overwrite").option("overwriteSchema", "true").format("delta").saveAsTable("house_housing_zipcode")
```
